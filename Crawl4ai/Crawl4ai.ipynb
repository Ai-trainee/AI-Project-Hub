```python
from crawl4ai import WebCrawler
```

```python
# Create an instance of WebCrawler
crawler = WebCrawler()
```

```python
# Warm up the crawler (load necessary models)
crawler.warmup()
```

```python
# Run the crawler on a URL
result = crawler.run(
    url = "https://venturebeat.com/ai/meet-the-new-most-powerful-open-source-ai-model-in-the-world-hyperwrites-reflection-70b/"
)

# Print the extracted content
print(result.markdown)
```
